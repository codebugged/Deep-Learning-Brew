```python
import os
```

```python
from imageai.Detection import ObjectDetection
```

```python
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
```

```python
model_path = 'resnet50_coco_best_v2.0.1.h5'
detector.setModelPath(model_path)
detector.loadModel()
```

```python
input_path = 'm.jpg'
output_path = 'm2.png'
```

```python
detections = detector.detectObjectsFromImage(input_image=input_path, output_image_path=output_path,minimum_percentage_probability=45)
```
